### 1. หารัฐที่มีจำนวนเครื่องบินขาออกมากที่สุดและน้อยที่สุด

````python
merged_table = flights.merge(airports,how='left',left_on='รหัสสนามบินต้นทาง',right_on='รหัสสนามบิน') #เชื่อมตาราง 2 ตาราง
merged_table
````

````python
merged_table
````

####เช็ค missing

````python
merged_table.dropna()
````

####Group by (pandas)

````python
merged_table.groupby('รัฐ').count() 
````

ดูGroup columns

````python
#จากโจทย์ถามหาขาออก
max=merged_table.groupby('รัฐ').count().sort_values(by=['รหัสสนามบินปลายทาง'],ascending=False) #Group columns ที่มีสนามบินปลายทางมากทึ่สุด
max.iloc[:1,4:5]
````

````python
#จากโจทย์ถามหาขาออก
min=merged_table.groupby('รัฐ').count().sort_values(by=['รหัสสนามบินปลายทาง'],ascending=False) #Group columns ที่มีสนามบินปลายทางน้อยทึ่สุด
min.iloc[:1,4:5]
````

###4.สายการบินอะไรมีความล่าช้ามากที่สุดและน้อยที่สุด

````python
merged_table['ความล่าช้าขาออก'].max() #ดู columns ความล่าช้าขาออกที่มีค่ามากที่สุด
````

````python
merged_table['ความล่าช้าขาเข้า'].max() #ดู columns ความล่าช้าขาออกที่มีค่ามากที่สุด
````

````python
max2_1 = merged_table.groupby('สายการบิน').max().sort_values(by=['ความล่าช้าขาออก'],ascending=False) # ความล่าช้าขาออกมากที่สุด 
max2_1.iloc[:1,4:5]
````

````python
max2_2 = merged_table.groupby('สายการบิน').max().sort_values(by=['ความล่าช้าขาเข้า'],ascending=False) # ความล่าช้าขาเข้ามากสุด 
max2_2.iloc[:1,5:6]
````
